```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load iris dataset from seaborn
df = sns.load_dataset('iris')
```

```python
# Shape of the dataset
print("Shape of the dataset:", df.shape)

# Column names
print("Column names:", df.columns.tolist())

# First 5 rows
print("\nFirst five rows:")
print(df.head())

# Info about dataset
print("\nInfo:")
print(df.info())

# Summary statistics
print("\nDescribe:")
print(df.describe())
```

## (a) Scatter Plot: Show relationships between two features

```python
# Pairplot to show relationships between all numerical features
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Iris Features', y=1.02)
plt.show()
```

## (b) Histograms: Show value distributions for each feature

```python
df.hist(edgecolor='black', figsize=(10, 8))
plt.suptitle('Histograms of Iris Features')
plt.tight_layout()
plt.show()
```

## (c) Box Plots: Identify outliers

```python
plt.figure(figsize=(10, 6))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f'Boxplot of {column} by Species')
plt.tight_layout()
plt.show()
```
