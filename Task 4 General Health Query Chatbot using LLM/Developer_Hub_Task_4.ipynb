```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# Load model and tokenizer
model_name = "mistralai/Mistral-7B-Instruct-v0.1"

!huggingface-cli login

print("Loading model...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

# Build a text-generation pipeline
chatbot = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Prompt Template
def build_prompt(user_query):
    return (
        "You are a helpful and friendly medical assistant. "
        "Answer general health questions in a simple, safe, and informative way. "
        "Avoid giving specific medical advice, diagnoses, or treatment plans. "
        "If the question is too complex or risky, advise the user to consult a doctor.\n\n"
        f"User: {user_query}\n"
        "Assistant:"
    )

# Safety check function
def is_safe_query(query):
    red_flags = [
        "should I take", "how much", "overdose", "suicide", "prescribe",
        "my child has a fever", "my baby", "chest pain", "emergency", "diagnose",
        "infection", "cancer", "stroke", "heart attack", "mental illness"
    ]
    return not any(flag in query.lower() for flag in red_flags)

# Main loop
def main():
    print("Health Chatbot (type 'exit' to quit)")
    while True:
        user_input = input("\nYou: ")
        if user_input.strip().lower() == "exit":
            break

        if not is_safe_query(user_input):
            print("Assistant: I'm sorry, but this question might be too sensitive or complex. "
                  "Please consult a qualified healthcare professional.")
            continue

        prompt = build_prompt(user_input)
        response = chatbot(prompt, max_new_tokens=200, do_sample=True, temperature=0.7)
        answer = response[0]['generated_text'].split("Assistant:")[-1].strip()
        print(f"Assistant: {answer}")

if __name__ == "__main__":
    main()
```
